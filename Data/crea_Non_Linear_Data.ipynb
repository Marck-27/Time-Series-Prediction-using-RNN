```python
# Cargamos Librerias
import matplotlib.pyplot as plt
import numpy as np 
from numpy.random import normal

import pandas as pd
```

## Construimos Datos con comportamiento no lineal:

```python
# Construimos datos con comportamiento No Lineal:
N=1000        

# =============== Construcción de Features ================
features = np.array( range(N) )


# ============ Construcción de labels en 2 pasos ============
 
# Paso 1. Generamos datos aleatorios con distribución normal:
media=0
sigma=1
rand_data  = normal(loc=media, scale=sigma, size=N) 

# Paso 2. Agregamos el patron no Lineal a los labels:
non_lin_patt = 5 * np.cos( 0.03 * features) + 5


labels = non_lin_patt + rand_data
labels.shape
```

```python
# Graficamos los datos:
plt.figure()
plt.figure(figsize=(15,5))

plt.plot( features , labels ,'o')
#plt.plot( features[100:200] , labels[100:200] ,'o',c='r')
plt.xlabel('Features')
plt.ylabel('Labels')
plt.title('Non Linear Data')

plt.show()
```

```python
# Concatenamos datos para pasarlos a un data frame de 2 columnas:

df = pd.DataFrame( {'x':features, 'y':labels} )

df.to_csv('non_lin_data_1D.csv', header=True, index=False)

```
